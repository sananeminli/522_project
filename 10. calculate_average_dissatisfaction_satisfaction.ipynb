```python
def is_functional_sentence(sentence):
  # Split the sentence into words
  words = sentence.split()

  # A functional sentence must have at least one subject and one verb
  if len(words) < 2:
    return False

  # Check if the sentence has a subject and a verb
  has_subject = False
  has_verb = False
  for word in words:
    if word in ['I', 'you', 'he', 'she', 'it', 'we', 'they']:
      has_subject = True
    elif word in ['am', 'is', 'are', 'was', 'were', 'be', 'being', 'been', 'do', 'does', 'did', 'have', 'has', 'had', 'will', 'would', 'shall', 'should', 'may', 'might', 'must']:
      has_verb = True
      
  # Return True if the sentence has both a subject and a verb
  return has_subject and has_verb
```

```python
satisfaction = []
dissatisfaction = []
neutral = []
freq = []
for x in range(len(features_list)):
    frequency = n_frequency = n_total_score = p_frequency = p_total_score = np_frequency = np_total_score = 0
    for y in range(len(feature_to_match)):
        if features_list[x] == feature_to_match[y]:
            frequency = frequency + 1
            if feature_sentiment[y] < 0:
                n_frequency = n_frequency + 1
                n_total_score = n_total_score + abs(feature_sentiment[y])
            elif feature_sentiment[y] > 0:
                p_frequency = p_frequency + 1
                p_total_score = p_total_score + feature_sentiment[y]
            elif feature_sentiment[y] == 0:
                np_frequency = np_frequency + 1
                np_total_score = np_total_score + 1
    freq.insert(x, frequency)
    
    if n_frequency > 0:
        n_avg = float(n_total_score*weight[x])/frequency
        dissatisfaction.insert(x, n_avg)
    elif n_frequency == 0:
        dissatisfaction.insert(x, 0)
        
    if p_frequency > 0:
        p_avg = float(p_total_score*weight[x])/frequency
        satisfaction.insert(x, p_avg)
    elif p_frequency == 0:
        satisfaction.insert(x, 0)
    
    if np_frequency > 0:
#         print(np_frequency)
        np_avg = float(np_total_score*weight[x])/frequency
        neutral.insert(x, np_avg)
    elif np_frequency == 0:
        neutral.insert(x, 0)
```
