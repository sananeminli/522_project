```python
import nltk
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
data = pd.read_csv('cleaned_tagged_sentences.csv') 
df = pd.DataFrame(data, columns=['r_id', 's_id', 'app_name','app_id', 'date', 'title', 'review', 'pos', 'tagged_sentences', 'new_tagged_sentences', 'rating'])
```

```python
sent_list = df['new_tagged_sentences'].tolist()
```

```python
for x in range(len(sent_list)):
    if re.search('costumer service', sent_list[x]):
        print("Old:", sent_list[x])
        string = sent_list[x]
        sent_list[x] = string.replace('costumer service', 'customer service')
        print("New:",sent_list[x])
    elif re.search('customer support', sent_list[x]):
        print("Old:", sent_list[x])
        string = sent_list[x]
        sent_list[x] = string.replace('customer support', 'customer service')
        print("New:",sent_list[x])
    elif re.search('call center', sent_list[x]):
        print("Old:", sent_list[x])
        string = sent_list[x]
        sent_list[x] = string.replace('call center', 'customer service')
        print("New:",sent_list[x])  
    elif re.search('support team', sent_list[x]):
        print("Old:", sent_list[x])
        string = sent_list[x]
        sent_list[x] = string.replace('support team', 'customer service')
        print("New:",sent_list[x]) 
    elif re.search('customer care', sent_list[x]):
        print("Old:", sent_list[x])
        string = sent_list[x]
        sent_list[x] = string.replace('customer care', 'customer service')
        print("New:",sent_list[x]) 
        
    if re.search('payment option', sent_list[x]):
        print("Old:", sent_list[x])
        string = sent_list[x]
        sent_list[x] = string.replace('payment option', 'payment method')
        print("New:", sent_list[x])
```

```python
# Getting bigrams 
vectorizer = CountVectorizer(ngram_range =(2, 2))
X1 = vectorizer.fit_transform(sent_list) 
features = (vectorizer.get_feature_names_out())
print("\n\nX1 : \n", X1.toarray())
```

```python
# Applying TFIDF
vectorizer = TfidfVectorizer(ngram_range = (2, 2))
X2 = vectorizer.fit_transform(sent_list)
scores = (X2.toarray())
print("\n\nScores : \n", scores)
```

```python
sums = X2.sum(axis = 0)
data1 = []
for col, term in enumerate(features):
    data1.append( (term, sums[0, col] ))
ranking = pd.DataFrame(data1, columns = ['term', 'rank'])
words = (ranking.sort_values('rank', ascending = False))
print ("\n\nWords : \n", words.head(100))
```

```python
features = words['term'].to_list()
weight = words['rank'].to_list() 
```

# Assign weight to my extracted features

```python
dataF = pd.read_csv('extracted_features.csv') 
dfF = pd.DataFrame(dataF, columns=['extracted_features', 'meaningful'])
```

```python
dfF['extracted_features']
```

```python
dfF.dropna(subset = ['extracted_features'], inplace=True)
```

```python
extracted_features = dfF['extracted_features'].to_list()
```

```python
# weight = []
        
table = {"extracted_features": [], "meaningful": [], "weight": []}
counter = 0

for x in range(len(extracted_features)):
    print("-------feature", extracted_features[x])
    for y in range(len(features)):
        if extracted_features[x] == features[y]:
            print("From my features:", extracted_features[x])
            print("Matched feature with:", features[y], "|| With weight::", weight[y])
            table["extracted_features"].append(extracted_features[x])
            table["meaningful"].append(dfF['meaningful'][x])
            table["weight"].append(weight[y])
```

```python
df=pd.DataFrame.from_dict(table,orient='index').transpose()
df.to_csv('features_with_weight.csv', index = False)
```
